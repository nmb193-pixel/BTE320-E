```python
sales = float(input("Enter Sales: "))
cogs = sales * 0.40
gross_profit = sales - cogs
tax_rate = float(input("Enter Tax Rate (%): "))
tax_amount = gross_profit * (tax_rate / 100)
net_profit = gross_profit - tax_amount
print(f"Net Sales: ${sales}")
print(f"Cost of Goods Sold: ${cogs}")
print(f"Gross Profit: ${gross_profit}")
print(f"Tax Amount: ${tax_amount}")
print(f"Net Profit: ${net_profit}")
```

```python
number1 = float(input("Number 1: "))
sign = input ("Operator: ")
number2 = float(input("Number 2: "))
if sign == "+":
  print(number1 + number2)
elif sign == "-":
  print(number1 - number2)
elif sign == "*":
  print(number1 * number2)
elif sign == "/":
    print(number1 / number2)
elif sign == "^":
      print(number1 ** number2)
elif sign == "%":
        print(number1 % number2)
else:
    print("Invalid Operator")
```

```python
base=5
rate=2.5
min_fee = 10
max_fee = 20
for hours in range(1,9):
    fee=base+(rate*hours)
    if fee < min_fee:
        fee = min_fee
    if fee > max_fee:
        fee = max_fee
    print(hours, fee)
```

```python
ticketprice=10
defaultaudience=20

print("ad$ Profit")
for admoney in range(0,201,25):
  revenue = (defaultaudience + 2*admoney ** .5) * ticketprice
  costs = 200+admoney
  profit=revenue - costs
  print (admoney, profit)
```

```python
base = 5
rate = 2.5
min_fee = 10
max_fee = 20

hours = 1
while hours <= 8:
    fee = base + (rate * hours)

    if fee < min_fee:
        fee = min_fee
    if fee > max_fee:
        fee = max_fee

    print(hours, fee)
    hours += 1
```

```python
ticket_price = 10
base_people = 20
fixed_cost = 200

print("Ad$ Profit")

ads = 0
while ads <= 200:
    extra_people = 2 * (ads ** 0.5)
    total_people = base_people + extra_people

    revenue = total_people * ticket_price
    profit = revenue - fixed_cost - ads

    print(ads, round(profit, 1))

    ads += 25
```

```python
import random

play = input("Do you want to play Rock, Paper, Scissors? (yes/no): ")

while play == "yes":

    # user move
    user_move = input("Enter rock, paper, or scissors: ")

    # computer randomly chooses
    computer_move = random.choice(["rock", "paper", "scissors"])

    print("Computer chose:", computer_move)

    if user_move == computer_move:
        print("Tie!")

    elif user_move == "rock":
        if computer_move == "scissors":
            print("You win!")
        else:
            print("Computer wins!")

    elif user_move == "paper":
        if computer_move == "rock":
            print("You win!")
        else:
            print("Computer wins!")

    elif user_move == "scissors":
        if computer_move == "paper":
            print("You win!")
        else:
            print("Computer wins!")

    else:
        print("Invalid input.")

    # ask to play again
    play = input("Do you want to play again? (yes/no): ")

print("Game over.")
```
